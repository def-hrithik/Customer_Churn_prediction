## Visualize churn distribution




**Reasoning**:
Import the necessary libraries for plotting and create a count plot to visualize the distribution of the 'Churn' column.



```python
import matplotlib.pyplot as plt
import seaborn as sns

demographic_features = ['gender', 'SeniorCitizen', 'Partner', 'Dependents']

for feature in demographic_features:
    plt.figure(figsize=(6, 4))
    sns.countplot(x=feature, hue='Churn', data=df)
    plt.title(f'Churn Distribution by {feature}')
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.show()
```

## Summary:

### Data Analysis Key Findings

*   The dataset contains 7043 rows and 21 columns, with no missing values.
*   Most customers are on Month-to-month contracts, which exhibit a significantly higher churn rate (42.7%) compared to One year (11.3%) and Two year contracts (2.9%).
*   Customers with higher Monthly Charges are more likely to churn, particularly those with charges exceeding \$70.
*   Shorter customer tenure is strongly associated with a higher likelihood of churning.
*   Customers with Fiber optic internet service have a much higher churn rate (41.9%) than those with DSL (18.9%) or no internet service (7.4%).
*   Electronic check is the payment method with the highest churn rate (45.3%).

### Insights or Next Steps

*   Develop targeted retention strategies for customers on Month-to-month contracts and those with Fiber optic internet service, perhaps by offering incentives for longer commitments or addressing potential service quality issues.
*   Investigate the reasons behind the high churn rate associated with high Monthly Charges and the Electronic check payment method to identify and mitigate potential pain points for these customer segments.
